# How to build a super easy handwritten numbers identifier
## Spoiler: use cosine similarity

$$\vec{A} \cdot \vec{B} = A_xB_x+A_yB_y+A_zB_z $$

$$\vec{A} \cdot \vec{B} = AB\cos(\theta) $$

$$ \cos(\theta)=\frac{\sum_i^n A_i B_i }{\sqrt{\sum_j^n A_j^2} \sqrt{ \sum_k^n B_k^2 }} $$

```python
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
```

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784',version=1)
X=mnist.data
```

```python
def some_im(x):
  return x.reshape(28,28)
```

```python
fig, axes = plt.subplots(2,5,figsize=(10,5),
                        subplot_kw={'xticks':(),'yticks':()})
for ax,img in zip(axes.ravel(),X):
  ax.imshow(some_im(img)) 
```

```python
plt.imshow(some_im(X[1]))
```

```python
lista=[]
indices=[]
n=5000
j=1 #random.randint(n)
for i in range(n):
  pp=np.dot(X[j],X[i])
  x1=np.linalg.norm(X[j])
  x2=np.linalg.norm(X[i])
  pp=pp/(x1*x2)
  sigma=0.15
  pp=np.exp(-0.5*( (pp-1)/sigma)**2)
  lista.append(pp)
  if (pp>0.3):
    indices.append(i)
```

```python
x=np.arange(len(lista))
fig = px.bar(x=x,y=lista)
fig.show()
# All the bars in the plot correspond to zeros.
```

```python
X_out=[]
for i in indices:
  X_out.append(X[i])
```

```python
fig, axes = plt.subplots(10,12,figsize=(10,5),
                        subplot_kw={'xticks':(),'yticks':()})
for ax,img in zip(axes.ravel(),X_out):
  ax.imshow(some_im(img)) 
```
